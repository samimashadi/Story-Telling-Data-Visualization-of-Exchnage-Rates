# Storytelling Data Visualization on Exchange Rates

The data set we will be working with in  this project describes Euro daily exchange rates between 1999 and 2021. The euro (symbolized with €) is the official currency in most of the countries of the European Union. The data set can be found [here.](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020) 

Lets begin by exploring the data. 

```python
import pandas as pd

exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
exchange_rates.info()
```

We see that in this data set, there are 41 columns and there are 5,699 total entries. we are no columns that only have null values. Out of the 41 columns in this data set, 38 of them are object data type and 3 of them are float type. 

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
exchange_rates.head(10)
```

```python
euro_to_dollar = exchange_rates[['Time','US_dollar']].copy()
euro_to_dollar['US_dollar'].value_counts()
```

Since our focus on this project is on the exchange rate between the euro and the American dollar, we will isolate the columns 'Time' and 'US_dollar' from our data set. 

After going over the US_dollar column, we see that there are 62 entries where a '-' is used. we will drop all the rows where this happens. We will also convert the entire column into a float type. 

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

## Rolling Means

Lets plot the 'Time' and the 'US_dollar' columns on a graph.

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.show()
```

If we look at the lines shape, we see that there are many small wiggles. This is because they are a visual representation of the daily variation  in the exchange rate. The rate goes up and down, up and down again, day to day. The rate only shows clear upward or downward trends in the longer run (months or years).

Lets use the rolling mean (also known as the moving average) to see the long term trends of the exchange rate. We will create a graph with the rolling window of 30 days. 

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()

plt.plot(euro_to_dollar['Time'], euro_to_dollar['rolling_mean'])
plt.title('Rolling Window 30 days')
plt.show()
```

Lets now create a grid chart of different rolling windows to find out a general trend. we will use a rilling window of 7, 30, 50, 100, 356.


```python
plt.figure(figsize=(15,12))

plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.title('Original Values', weight = 'bold')

for i, rolling_mean in zip([2,3,4,5,6],[7,30,50,100,365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'].rolling(rolling_mean).mean())
    plt.title('Rolling Window: ' + str(rolling_mean), weight='bold')
    
plt.show()
```

## Euro-USD Exchange rate under last 3 US presidents

Using the data set we have, we will create an explanatory visualization of how the Euro and US dollar exchange rate fluctuated during the presidency of Geroge Bush (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021). 

First lets isolate the data we have with respect to time under each president. 

```python
BOT = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2001) & (euro_to_dollar['Time'].dt.year < 2021)]

bush = BOT.copy()[BOT['Time'].dt.year < 2009]
obama = BOT.copy()[(BOT['Time'].dt.year >= 2009) & (BOT['Time'].dt.year < 2017)]
trump = BOT.copy()[(BOT['Time'].dt.year >= 2017)]

print(BOT['US_dollar'].mean())
print(bush['US_dollar'].mean())
print(obama['US_dollar'].mean())
print(trump['US_dollar'].mean())
```

We have calculated the average exchange rate over the last 20 years. It is 1.22. Now we will create our graph.

```python
import matplotlib.style as style 
style.use('fivethirtyeight')

### creating the subplots
plt.figure(figsize=(12,6))
ax1 = plt.subplot(2,3,1)
ax2 = plt.subplot(2,3,2)
ax3 = plt.subplot(2,3,3)
ax4 = plt.subplot(2,1,2)
axes = [ax1,ax2,ax3,ax4]

### format all subplots

for ax in axes:
    ax.set_ylim(0.8,1.7)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(['1.0','1.2','1.4','1.6'], alpha = 0.3)
    ax.grid(alpha=0.5)
    
### subplot for bush
ax1.plot(bush['Time'],bush['rolling_mean'], color = '#FE420F')
ax1.set_xticklabels(['','2001','','2003','','2005','','2007','','2009'], alpha = 0.3)
ax1.text(731481, 1.92, 'BUSH', fontsize = 18, weight = 'bold', color='#FE420F')
ax1.text(731216, 1.8, '(2001-2009)', weight = 'bold', alpha = 0.3)

### subplot for Obama
ax2.plot(obama['Time'],obama['rolling_mean'], color = '#0000FF')
ax2.set_xticklabels(['','2009','','2011','','2013','','2015','','2017'], alpha = 0.3)
ax2.text(734303, 1.92, 'OBAMA', fontsize = 18, weight = 'bold', color='#0000FF')
ax2.text(734138, 1.8, '(2009-2017)', weight = 'bold', alpha = 0.3)

### subplot for Trump
ax3.plot(trump['Time'],trump['rolling_mean'], color = '#EE82EE')
ax3.set_xticklabels(['2017','','2018','','2019','','2020','','2021'], alpha = 0.3)
ax3.text(736950, 1.92, 'TRUMP', fontsize = 18, weight = 'bold', color='#EE82EE')
ax3.text(736876, 1.8, '(2009-2017)', weight = 'bold', alpha = 0.3)

### subplot for Bush Obama
ax4.plot(bush['Time'], bush['rolling_mean'], color = '#FE420F')
ax4.plot(obama['Time'], obama['rolling_mean'], color = '#0000FF')
ax4.plot(trump['Time'], trump['rolling_mean'], color = '#EE82EE')
ax4.grid(alpha = 0.5)
ax4.set_xticks([])

### adding title and subtitle 
ax1.text(730120,2.35, 'The average Euro-USD exchange rate over the last 20 years is 1.22', fontsize = 20, weight = 'bold')
ax1.text(730120, 2.20, 'Euro-USD exchange rate under the last 3 presidents', fontsize = 18)
plt.show()

```

We have created a explanatory Visualization of the Euro to US dollar exchnage rate. The following is a basic analysis. 

It seems that the during the era of George Bush, the general trend of the exchnage rate was that it was increasing. The average exchange rate during the the time Geroge Bush was president was 1.195, which is actually the second lowest out of the three presidents. This is because in the start of his era, the exchnage rate was very low. Even lower than 1 and after 2003 did it pass 1. However during his presidency, the exchnage rate also climbed to its highest reaching almost 1.6. 

Once Obama came into presidency, The general trend of the exchnage rate seemed to be going down. The average exchnage rate during obamas time was 1.284 which is the highest out of the three presidents. This is because in the start of Obama's presidency, the exchnage rate was high but by the time Obama left office, the exchnage rate had gone down almost to 1.0.

During Trumps presidency, the general trend was more or less static. Trump started his presidency with the exchange rate being just a littel bit above 1 and by the end it was at 1.2. The average exchange rate during his time was 1.143 which is actually the lowest. It reached a high of around 1.3 during 2018 and a low of about 1.1 in the middle of 2020.
